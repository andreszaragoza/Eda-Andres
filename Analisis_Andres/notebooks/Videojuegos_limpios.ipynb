#  Limpiar columnas JSON
## Detectar y manejar outliers en `playtime`, `ratings_count`, `reviews_count`
- Aplicar método de IQR para detección de outliers

```python
import ast

def extract_names(json_str):
    """Extrae los nombres de un JSON en formato string."""
    try:
        json_list = ast.literal_eval(json_str)  
        return [item['name'] for item in json_list] if isinstance(json_list, list) else None
    except (ValueError, SyntaxError):
        return None

df['released'] = pd.to_datetime(df['released'], errors='coerce')


df['stores'].fillna("No Disponible", inplace=True)  

def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

outlier_columns = ['playtime', 'ratings_count', 'reviews_count']
for col in outlier_columns:
    df = remove_outliers(df, col)
```
